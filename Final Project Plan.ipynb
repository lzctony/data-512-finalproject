The dataset consists of 5 files:
1. **schema.csv**: a CSV file with survey schema. This schema includes the questions that correspond to each column name in both the **multipleChoiceResponses.csv** and **freeformResponses.csv**.
2. **multipleChoiceResponses.csv**: Respondents' answers to multiple choice and ranking questions.
3. **freeformResponses.csv**: Respondents' freeform answers to Kaggle's survey questions.
4. **conversionRates.csv**: Currency conversion rates to USD (accessed from the R package "quantmod" on September 14, 2017)
5. **RespondentTypeREADME.txt**: This is a schema for decoding the responses in the "Asked" column of the **schema.csv** file.
